# Percobaan Model

```python
from scipy.sparse import csr_matrix
```

```python
skincare_sparse = csr_matrix(skincare_pivot)
```

```python
type(skincare_sparse)
```

```python
# import algoritma dari Nearest Neighbors 

from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(skincare_sparse)
```

```python
index_to_access = 7  # Ubah sesuai dengan indeks yang bermasalah

# Print ukuran DataFrame dan nilai indeks yang akan diakses
print(f"Ukuran DataFrame: {skincare_pivot.shape}")
print(f"Indeks yang diakses: {index_to_access}")

# Coba akses data
try:
    data = skincare_pivot.iloc[index_to_access]
    print(data)
except IndexError as e:
    print(f"Kesalahan: {e}")
    print(f"Indeks {index_to_access} berada di luar batas yang valid.")
```

# Pengujian Model

```python
def recommend_skincare(skincare_name):
    skincare_id = np.where(skincare_pivot.index == skincare_name)[0][0]
    distance, suggestion = model.kneighbors(skincare_pivot.iloc[skincare_id,:].values.reshape(1,-1), n_neighbors=4 )
    
    for i in range(len(suggestion)):
            skincare = skincare_pivot.index[suggestion[i]]
            for j in skincare:
                if j == skincare_name:
                    print(f"You searched '{skincare_name}'\n")
                    print("The suggestion skincare are: \n")
                else:
                    print(j)
```

```python
skincare_name = "White-Expert-2in1-Cleanser-&-Toner"
recommend_skincare(skincare_name)
```
